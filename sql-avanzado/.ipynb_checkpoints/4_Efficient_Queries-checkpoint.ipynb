# Queries más eficientes
En ocasiones, la eficiencia de nuestra *query* no tendrá gran importancia, por ejemplo, si está diseñada para correr una única vez, o sobre *datasets* pequeños. Pero, en otros casos, si ésta se ejecutará muchas veces o sobre *datasets* de gran tamaño, si se utiliza para proveer información a una página web, por ejemplo, será vital que lo haga de manera eficiente, de manera de asegurar su usabilidad (y no dejar a un usuario esperando mucho tiempo para poder visualizar los datos, por ejemplo).

Algunas funciones que nos permiten evaluar la efectividad de nuestras *queries* (en función de su tamaño y tiempo de ejecución), son:


## Estrategias Efectividad
**1. Seleccionar SÓLO las columnas deseadas:** esto es, no dejar para la tabla final todas las columnas, sobre todo **evitar aquellas que incluyan texto**, ya que estas suelen ocupar grandes tamaños de memoria.

**2. Leer menos datos:** esto es, evitar la lectura de diferentes columnas lo mayor posible, por ejemplo, si hay variables relacionadas y puedo evitar utilizar una para *SELECT* y otra para *WHERE*, **evitarlo**.

**3. Evitar N:N JOINs:** son aquellas donde un grupo de filas de una tabla puede estar vinculado con un grupo de filas de otra, por lo que, de realizar un *JOIN* se incorporará una gran cantidad de filas en la tabla resultante


Este era el caso de un *JOIN 1:1*


Este es aquel donde, una fila de una tabla puede +matchear* con varias de la otra *JOIN N:1*


Y en este último caso, diferentes filas pueden matchear con varias de la otra tabla *JOIN N:N*


Veamos un ejemplo, donde se mejora una *query* que, en primer lugar, ocupaba gran tamaño, producto de realizar *JOIN* de gran extensión. Luego, se reduce el tamaño de este *JOIN*, por medio de preconsultas que reducen los datos a combinar.

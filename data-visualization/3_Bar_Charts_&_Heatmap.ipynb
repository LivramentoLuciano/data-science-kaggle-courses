# Gráfico de Barras y Calor

## Gráfico de barras
Este tipo de gráficos, al igual que el de líneas es de una ejecución muy sencilla, con una simple línea de código.

```python
# Configuraciones inciales
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

# Datos con los que trabajaremos 
# Demoras de Aerolineas, año 2015
flight_data = pd.read_csv('assets/input/flight_delays.csv', index_col='Month')
```

```python
# inspecciono data
flight_data.head()
```

```python
# gráfico de barras
plt.figure(figsize=(16,6))
plt.title('Demoras de "NK" durante año 2015 (promedio mensual)')

sns.barplot(x=flight_data.index, y=flight_data['NK'])

plt.xlabel('Mes')
plt.ylabel('Demora promedio (en minutos)')
```

## Mapa de Calor
Estos nos permite visualizar patrones dentor de un gran conjunto de datos de manera muy simple, empleando un código de colores

```python
# Mapa de Calor
# gráfico de barras
plt.figure(figsize=(16,6))
plt.title('Demoras de las Aerolíneas durante año 2015 (promedio mensual)')

sns.heatmap(data=flight_data, annot=False)

plt.xlabel('Aerolínea')
plt.ylabel('Mes')
```

## Práctica
Utilizando los datos provistos en el dataset de *IGN Reviews*, se desea analizar la valoración que tienen los distintos géneros de juegos digitales y en sus diferentes plataformas, para así tomar una decisión sobre un futuro videojuego a lanzar.

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

# cargo data
ign_data = pd.read_csv('assets/input/ign_scores.csv', index_col='Platform')
ign_data
```

### 1. ¿Cuál es el máximo puntaje recibido por un juego de PC?

```python
highest_pc_score = ign_data.loc['PC'].max()
highest_pc_score
```

### 2. ¿Cuál es el género que obtuvo el peor puntaje de la plataforma Playstation Vita?

**NOTA:** va a requerir encontrar un método que me permita acceder al nombre de la columna, que presenta el valor mínimo (en este caso, pero seguramente también exista para máximo) dentro de una serie de valores.

La función encontrada para esto es `idxmin()` [(info)](https://pandas.pydata.org/docs/reference/api/pandas.Series.idxmin.html). Esta devuelve directamente el *label* de la posición que presente el valor mínimo dentro de la *Series* (también funciona en *Dataframes*)

```python
# ColumnLabel que presenta el valor Mínimo
worst_vita_genre = ign_data.loc['PlayStation Vita'].idxmin()
worst_vita_genre
```

### 3. ¿Cuál es la plataforma adecuada para tu juego?
Teniendo en cuenta que tu juego favorito es "Mario Karting" de la Nintendo Wii, deseas desarrollar un videojuego de carreras para esta misma plataforma.

Verificar si esta plataforma es la mejor elección para un juego de este género (*racing*). Crear un gráfico de barras que muestre el puntaje recibido para este género en cada plataforma.

```python
plt.figure(figsize=(16,6))
plt.title('Puntaje por plataforma, para el género "Racing"')

sns.barplot(x=ign_data.index, y=ign_data['Racing'])

plt.xlabel('Plataforma')
plt.ylabel('Puntaje')
```

Se observa que la plataforma **Nintendo Wii, no será una buena elección** si se desea desarrollar un videojuego de carreras. En su lugar, **se podría seleccionar a la Xbox One.**

### 4. ¿Cuál sería la mejor decisión (plataforma/género)?
Se desiste de desarrollar un videojuego de carreras para la Wii, no obstante, se desea aún desarrollar un videojuego. Para esto, se pretende visualizar la puntuación por género para cada plataforma.

Crear un mapa de calor que nos permita encontrar fácilmente cuál combinación de género y plataforma obtuvo el mejor puntaje.

```python
plt.figure(figsize=(16,6))
plt.title('Puntaje por plataforma, para cada género')

sns.heatmap(data=ign_data, annot=True)

plt.xlabel('Género')
plt.ylabel('Plataforma')
```

A partir del gráfico podemos ver rápidamente que los videojuegos mejor valorados son los de *Simulation* dentro de la plataforma *PlayStation 4*.

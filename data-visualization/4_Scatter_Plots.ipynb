# Gráficos de Dispersión
Estos gráficos nos ayudan a entender la relación entre variables (2, en principio, pero veremos la forma de incluir una tercera).

Nuevamente, para obtener este gráfico sólo hará falta 1 línea de código:

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

# datos a utilizar
insurance_data = pd.read_csv('assets/input/insurance.csv')
insurance_data
```

```python
plt.figure(figsize=(8,4))
plt.title('Relación BMI-Cargos de Seguro en una población')

# grafico de dispersion
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'])

plt.xlabel('Indice de Masa Corporal')
plt.ylabel('Costo')
```

## Línea de Dispersión
Del anterior ejemplo, podemos concluir que existe una relación directamente proporcional entre el *BMI* y los costos del seguro, pero para visualizar la misma con mayor seguridad, podemos incorporar al gráfico una línea de regresión.

```python
plt.figure(figsize=(8,4))
plt.title('Relación BMI-Cargos de Seguro en una población')

# grafico de dispersion simple con linea de regresión
sns.regplot(x=insurance_data['bmi'], y=insurance_data['charges'])

plt.xlabel('Indice de Masa Corporal')
plt.ylabel('Costo')
```

## Gráfico de Dispersión de 3 variables
En un gráfico de dispersión, existe la posibilidad de incorporar una tercera variable que ayude aún más a entender la relación entre las mismas (creo que deberá ser una variable categórica), incorporando una codificación por colores en nuestros datos graficados.

Por ejemplo, siguiendo con el ejemplo anterior, se podría incluir la condición *smoker* de la persona y así saber cómo afecta ésta en el costo del seguro, sumado al ya ponderado valor de *bmi*. Para esto se utiliza el parámetro *hue* del *scatterplot*.

```python
plt.figure(figsize=(8,4))
plt.title('Relación BMI-Cargos, contemplando condición de "Fumador"')

# grafico de dispersion de 3 variables (color coded)
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'], hue=insurance_data['smoker'])

plt.xlabel('Indice de Masa Corporal')
plt.ylabel('Costo')
```

Incorporando esta variable en el gráfico (color) podemos detectar que en los **no fumadores, el aumento de masa corporal no implica un gran aumento de los cargos del seguro**, mientras que, en el caso de los **fumadores, estos cargos se ven notablemente incrementados al subir el mencionado índice**.

Análogamente al caso del *scatterplot* simple, se pueden incorporar líneas de regresión en un gráfico de 3 variables, de la siguiente manera (*lmplot*).

**IMPORTANTE:** este gráfico utiliza una manera diferente de describir los árgumentos, esto es, indicando los nombres de las columnas deseadas y no las columnas como tales (como era en el caso del *scatterplot*).

```python
plt.figure(figsize=(12,6))

# grafico de dispersion con 3 variables (color coded) + Lineas de regresion
sns.lmplot(x='bmi', y='charges', hue='smoker', data=insurance_data)

plt.title('Relación BMI-Cargos contemplando "Fumador", con Regresión')
plt.xlabel('Indice de Masa Corporal')
plt.ylabel('Costo')
```

## Gráfico de Dispersión Categórico
Este nuevo gráfico permite **relacionar 2 variables, siendo una de ellas categórica** (discreta).

```python
plt.figure(figsize=(8,4))
plt.title('Relación Fumador-Cargos, Dispersión discreta')

# grafico de dispersion de variable discreto 
sns.swarmplot(x=insurance_data['smoker'], y=insurance_data['charges'])

plt.xlabel('Fumador')
plt.ylabel('Costo')
```

De este nuevo gráfico, podemos concluir, por ejemplo, que en general, los "no fumadores" pagarán un menor precio de seguro que los "fumadores".

Para practicar, realizo otro gráfico, ahora con la relación entre la cantidad de hijos y el costo del seguro.

```python
plt.figure(figsize=(8,4))
plt.title('Relación Fumador-Cargos, Dispersión discreta')

# grafico de dispersion de variable discreto 
sns.swarmplot(x=insurance_data['children'], y=insurance_data['charges'])

plt.xlabel('Fumador')
plt.ylabel('Costo')
```

## Práctica
Se desea diseñar un nuevo producto en una empresa de dulces. Para ello, se deberá realizar un análisis de los datos disponibles de encuestas a la población sobre sus gustos y preferencias en cuanto a este tipo de productos.

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

candy_data = pd.read_csv('assets/input/candy.csv', index_col='id')
candy_data.head()
```

### 1. Importancia del azúcar
¿Las personas tienden a tener preferencia por dulces con mayor contenido de azúcar?

```python
plt.figure(figsize=(8,4))
plt.title('Relación entre "sugarpercent" y "winpercent"')

sns.regplot(x=candy_data['sugarpercent'], y=candy_data['winpercent'])
plt.xlabel('% Azúcar')
plt.ylabel('% Preferencia del consumidor')
```

Del gráfico se observa que la relación entre el contenido de azúcar y el gusto de los consumidores es **levemente positiva**. Por tanto, podemos decir que tienden a prefereir levemente los dulces que tienen un mayor contendido de azúcar.

### 2. Análisis del chocolate
Ahora analizaremos las preferencias del público respecto a los dulces que contienen chocolate.

Comenzamos realizando un gráfico que meustre la relación "*precio-preferencia*", discriminando, también, por contenido de chocolate o no.

```python
plt.figure(figsize=(8,4))

sns.lmplot(x='pricepercent', y='winpercent', hue='chocolate', data=candy_data)

plt.title('Relación entre "price" y "winpercent"')
plt.xlabel('Precio')
plt.ylabel('% Preferencia del consumidor')
```

Del gráfico anterior, puede concluir que hay una relación positiva en cuanto al precio y la preferencia del consumidor, entre los dulces de chocolate. Por tanto, los **chocolates caros gustan más que los baratos**.

Por otro lado, en los dulces sin chocolate, la relación es inversa, por lo que los **dulces sin chocolate, son preferidos cuando son más baratos.**

Finalmente, utilizo un diagrama categórico para identificar la relación *chocolate-preferencia*.

```python
plt.figure(figsize=(8,4))

sns.swarmplot(x=candy_data['chocolate'], y=candy_data['winpercent'])

plt.title('Relación entre "chocolate" y "winpercent"')
plt.xlabel('Chocolate')
plt.ylabel('% Preferencia del consumidor')
```

De este gráfico concluimos en que los chocolates suelen ser de mayor preferencia que los dulces sin chocolate.

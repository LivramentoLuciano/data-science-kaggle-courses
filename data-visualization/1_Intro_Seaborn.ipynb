# Introducción Seaborn
En este curso se aprenderá a realizar gráficos de gran utilidad de una manera muy sencilla, por medio de `Seaborn`. Esta librería nos permite realizar gráficos, en muchas ocasiones, con tan solo 1 línea de código.

Para ejemplificar esto, graficaremos los datos provenientes de un dataset de *FIFA*, con información sobre la posición histórica en el ranking de algunos de los principales países en el ámbito del fútbol.

## Práctica

```python
# Librerias a utilizar
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

# cargo los datos
fifa_data_filepath = 'assets/input/fifa.csv'
fifa_data = pd.read_csv(fifa_data_filepath, index_col="Date", parse_dates=True)
```

```python
# inspecciono los datos
fifa_data.head()
```

```python
# seteo tamaño de la figura
plt.figure(figsize=(12,6))

# Grafico ejemplo: linea
sns.lineplot(data=fifa_data)
```

```python
plt.figure(figsize=(12,6))

# solo para practicar
# puedo graficar solo algunos
sns.lineplot(data=fifa_data[["ARG", "BRA"]])
```

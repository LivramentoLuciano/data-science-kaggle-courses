# Seleccin de Grficos y Estilo

En la siguiente imagen se puede apreciar una sencilla manera de seleccionar cul ser el grfico correcto a utilizar, dependiendo de las caractersticas de nuestro problema y lo que deseamos visualizar.



Se han divido los tipos de grficos en 3 categoras:
- **Tendencias:**
    - *lineplot:* son los mejores para representar tendencias a lo largo del tiempo. Adems, permiten visualizar informacin de diferentes grupos al mismo tiempo.
- **Relaciones:**
    - *barplot:* tiles para comparar cantidades correspondientes a diferentes grupos.
    - *heatmpa:* permiten encontrar patrones (por cdigo de colores) en tablas de nmeros.
    - *scatterplot:* muestran la relacin entre 2 variables continuas. Incluso, permiten incorporar una tercera variable categrica en la relacin (mediante cdigo de colores).
    - *regplot:* incluye una lnea de regresin para mejor comprensin de un *scatter plot*.
    - *lmplot:* idem al anterior, en *scatter plots* que estn divididos en grupos de colores por una variable categrica.
    - *swarmplot:* son *scatter plots* categricos, representan la relacin entre una variable categrica y una continua.
- **Distribucin:**
    - *distplot:* histograma, muestra la distribucin de una variable numrica (frecuencia, incidencias de un valor).
    - *kdeplot:* densidad de distribucin de una variable numrica (una versin suavizada del histograma).
    - *jointplot:* combinacin de 2 grficos de densidad, incluyendo las densidades por separado y la combinacin de ocurrencias de ambas.

## Estilos
Segn hemos visto hasta ahora, *seaborn* nos permite obtener grficos con una muy buen interfaz, no obstante, adems, podremos modificar el estilo aplicado a nuestro grfico con una simple lnea de cdigo `sns.set_style()`.

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

spotify_data = pd.read_csv('assets/input/spotify.csv', index_col="Date", parse_dates=True)
plt.figure(figsize=(8,4))
sns.lineplot(data=spotify_data)
```

```python
sns.set_style('dark')

plt.figure(figsize=(8,4))
sns.lineplot(data=spotify_data)
```

Los estilos disponibles: 
- white
- dark
- whitegrid
- darkgrid
- ticks

```python
sns.set_style('whitegrid')

plt.figure(figsize=(8,4))
sns.lineplot(data=spotify_data)
```

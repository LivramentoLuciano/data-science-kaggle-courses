# Selección de Gráficos y Estilo

En la siguiente imagen se puede apreciar una sencilla manera de seleccionar cuál será el gráfico correcto a utilizar, dependiendo de las características de nuestro problema y lo que deseamos visualizar.



Se han divido los tipos de gráficos en 3 categorías:
- **Tendencias:**
    - *lineplot:* son los mejores para representar tendencias a lo largo del tiempo. Además, permiten visualizar información de diferentes grupos al mismo tiempo.
- **Relaciones:**
    - *barplot:* útiles para comparar cantidades correspondientes a diferentes grupos.
    - *heatmpa:* permiten encontrar patrones (por código de colores) en tablas de números.
    - *scatterplot:* muestran la relación entre 2 variables continuas. Incluso, permiten incorporar una tercera variable categórica en la relación (mediante código de colores).
    - *regplot:* incluye una línea de regresión para mejor comprensión de un *scatter plot*.
    - *lmplot:* idem al anterior, en *scatter plots* que estén divididos en grupos de colores por una variable categórica.
    - *swarmplot:* son *scatter plots* categóricos, representan la relación entre una variable categórica y una continua.
- **Distribución:**
    - *distplot:* histograma, muestra la distribución de una variable numérica (frecuencia, incidencias de un valor).
    - *kdeplot:* densidad de distribución de una variable numérica (una versión suavizada del histograma).
    - *jointplot:* combinación de 2 gráficos de densidad, incluyendo las densidades por separado y la combinación de ocurrencias de ambas.

## Estilos
Según hemos visto hasta ahora, *seaborn* nos permite obtener gráficos con una muy buen interfaz, no obstante, además, podremos modificar el estilo aplicado a nuestro gráfico con una simple línea de código `sns.set_style()`.

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

spotify_data = pd.read_csv('assets/input/spotify.csv', index_col="Date", parse_dates=True)
plt.figure(figsize=(8,4))
sns.lineplot(data=spotify_data)
```

```python
sns.set_style('dark')

plt.figure(figsize=(8,4))
sns.lineplot(data=spotify_data)
```

Los estilos disponibles: 
- white
- dark
- whitegrid
- darkgrid
- ticks

```python
sns.set_style('whitegrid')

plt.figure(figsize=(8,4))
sns.lineplot(data=spotify_data)
```

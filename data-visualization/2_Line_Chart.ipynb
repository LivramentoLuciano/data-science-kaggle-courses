# Gráficos de Línea
Como ya vimos en la clase introductoria, un gráfico de línea es muy sencillo de realizar, tanto así que sólo ocupa 1 línea de código (*sns.lineplot*).

Ahora, en adición al gráfico en sí, hay algunas otras cuestiones que se pueden configurar de nuestra figura, como el título o etiqueta de los ejes.

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

spotify_data = pd.read_csv('assets/input/spotify.csv', index_col='Date', parse_dates=True)
```

```python
spotify_data.head()
```

```python
plt.figure(figsize=(8,4))
plt.title('Reproducciones a nivel global 2017-2018')

sns.lineplot(data=spotify_data)
plt.xlabel('Fecha')
```

## Gráfico Subsets
Como hemos visto, también en la anterior presentación, se pueden granficar, si así se desea, ciertas columnas en particular. Tan solo se deberá indicar cuáles se requieren.

```python
plt.figure(figsize=(8,4))
plt.title('Reproducciones a nivel global ("Shape Of You", "Despacito") 2017-2018')

# Grafico de subsets (individualmente)
sns.lineplot(data=spotify_data['Shape of You'], label='Shape of You')
sns.lineplot(data=spotify_data['Despacito'], label='Despacito')
plt.xlabel('Fecha')
```

La anterior solución es una posibilidad, pero (creo yo) que es mejor si se seleccionan todos las columnas a graficar en una sola instrucción, ya que, además, me evita tener que indicar las *labels* y no genera el error visual que produce la anterior que coloca un *label* en el eje de abcisas que genera confusión.

```python
plt.figure(figsize=(8,4))
plt.title('Reproducciones a nivel global ("Shape Of You", "Despacito") 2017-2018')

# Grafico de subsets (1 sola instrucción)
sns.lineplot(data=spotify_data[['Shape of You', 'Despacito']])
plt.xlabel('Fecha')
```

# Práctica
Realizamos una práctica, basándonos en un dataset que contiene información sobre los visitantes mensuales a 4 museos de la ciudad de Los Angeles.

```python
museum_data = pd.read_csv('assets/input/museum_visitors.csv', index_col='Date', parse_dates=True)
museum_data.head()
```

### 1. Visitantes durante Evento
Desde el *Firehouse Museum* desean convencer a los administradores de otorgar un mayor presupuesto para lanzar un evento que ellos afirman que, en ocasiones anteriores, ha convocado a una enorme cantidad de visitantes (año 2014). 

Realizar un gráfico que demuestre si este incremento fue tal y como se afirma, comparado con los visitantes regulares de todos los museos.

```python
plt.figure(figsize=(12,6))
plt.title('Visitantes mensuales de cada museo (2014-2016)')
plt.xlabel('Fecha')

sns.lineplot(data=museum_data)
```

### 2. Temporadas en el Avila Adobe
se desea predecir las temporadas de mayor/menor convocatoria de visitantes al museo, de manera tal de poder gestionar mejor la cantidad de empleados que se requerirán.

Crear un gráfic de línea que muestre cómo evoluciona la cantidad de visitantes en este museo a lo largo del tiempo.

```python
plt.figure(figsize=(14,4))
plt.title('Visitantes mensuales del "Avila Adobe" (2014-2016)')
plt.xlabel('Fecha')

sns.lineplot(data=museum_data['Avila Adobe'])
```

Se observa que en los períodos de Marzo-Agosto (primavera y verano) se presentan las mayores cantidades de visitantes, mientras que en el período Septiembre-Febrero (otoño e invierno), este número disminuye. Por tanto, este museo, podría verse beneficiado de contratar mayor personal para la época de Primavera/Verano donde la atención es máxima.

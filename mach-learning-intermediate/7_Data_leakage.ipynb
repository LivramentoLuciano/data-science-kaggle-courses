# Data Leakage

Es uno de los conceptos más importantes a tener en cuenta al practicar *data science* ya que, de ocurrir, puede arruinar nuestros modelos de manera sútil pero peligrosa. 

Ocurre cuando nuestros **datos de entrenamiento contienen información sobre el *target*, pero esta no estará disponible al momento de realizar la predicción**. Esto genera que veamos una **gran performance en el *training_set* pero muy pobre en *production***, volviendo a nuestro **modelo inútil**.

Hay 2 tipos, principalmente:
- **Target leakage:** cuando incluimos en el *predictor* data que no va a estar disponible al momento de hacer las predicciones. Se debe pensar en éste en términos cronológicos de disponibilidad de la información. Ejemplo, se quiere predecir si una persona va a contraer neumonía, una de las *features* indica si "took_meds*, la cual será True una vez que la persona haya enfermado efectivamente. Por tanto, si entrenamos a nuestro modelo incluyendo este dato, encontrará una relación directa entre esta variable y el *target* incurriendo en un error por leakage. En resumen, se debe evitar incluir parámetros que no serán creados o cambiarán hasta el momento en que el evento a predecir ocurra.


- **Train-test contamination:** ocurre cuando **contaminamos los datos de entrenamiento con los de validación**, obteniendo una buena performance de *validation* ya que el modelo fue entrenado con data que estaba incluida en dicho set, por tanto no es data nueva, pero **pobre en producción.** Un ejemplo de ocurrencia de este error sería si realizáramos una imputación de valores *missing* en nuestros datos, antes de separarlos *train_test_split*, por tanto, la imputación estará basada en información tanto de *train* como de *validation* (ergo, info de entrenamiento contaminada)

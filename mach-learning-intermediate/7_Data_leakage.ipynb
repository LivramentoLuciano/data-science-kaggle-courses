# Data Leakage

Es uno de los conceptos ms importantes a tener en cuenta al practicar *data science* ya que, de ocurrir, puede arruinar nuestros modelos de manera stil pero peligrosa. 

Ocurre cuando nuestros **datos de entrenamiento contienen informacin sobre el *target*, pero esta no estar disponible al momento de realizar la prediccin**. Esto genera que veamos una **gran performance en el *training_set* pero muy pobre en *production***, volviendo a nuestro **modelo intil**.

Hay 2 tipos, principalmente:
- **Target leakage:** cuando incluimos en el *predictor* data que no va a estar disponible al momento de hacer las predicciones. Se debe pensar en ste en trminos cronolgicos de disponibilidad de la informacin. Ejemplo, se quiere predecir si una persona va a contraer neumona, una de las *features* indica si "took_meds*, la cual ser True una vez que la persona haya enfermado efectivamente. Por tanto, si entrenamos a nuestro modelo incluyendo este dato, encontrar una relacin directa entre esta variable y el *target* incurriendo en un error por leakage. En resumen, se debe evitar incluir parmetros que no sern creados o cambiarn hasta el momento en que el evento a predecir ocurra.


- **Train-test contamination:** ocurre cuando **contaminamos los datos de entrenamiento con los de validacin**, obteniendo una buena performance de *validation* ya que el modelo fue entrenado con data que estaba incluida en dicho set, por tanto no es data nueva, pero **pobre en produccin.** Un ejemplo de ocurrencia de este error sera si realizramos una imputacin de valores *missing* en nuestros datos, antes de separarlos *train_test_split*, por tanto, la imputacin estar basada en informacin tanto de *train* como de *validation* (ergo, info de entrenamiento contaminada)
